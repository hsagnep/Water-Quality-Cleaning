```python
import pandas as pd
import datetime
import numpy as np
```

<h1>Data Cleaning</h1>

```python
#India Affected Water Quality Areas data: Government data related to the water quality of India
df = pd.read_csv("/Users/hervinsagnep/Downloads/IndiaAffectedWaterQualityAreas.csv",engine='python')
```

```python
#Showing first 5 rows of data
df.head()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
#Checking data types of the columns
df.info()
```

```python
#Changing year from string into timeseries
change_datetime = lambda d: datetime.datetime.strptime(d,'%d/%m/%y')
df['Year'] = df['Year'].apply(change_datetime)
```

```python
#Setting index to Year 
df.set_index('Year',inplace=True,drop=True)
```

```python
#Final Dataframe
df.head()
```

```python
#Indian water quality data: historical data on pollution levels and temperature across India
second_df = pd.read_csv("/Users/hervinsagnep/Downloads/water_dataX.csv",engine='python')
```

```python
second_df.head()
```

```python
#Descending year order
second_df = second_df[:-1]
```

```python
#Checking for NaN 
second_df.isnull().sum()
```

```python
#Checking data types of the columns
second_df.info()
```

```python
#Replacing 'NAN' and empty values with NaN
second_df.replace(['NAN',' '],np.nan,inplace=True)
```

```python
#Real number of NaN values
second_df.isnull().sum()
```

```python
#List of column names
column_names = list(second_df)
```

```python
#Changing columns 4 - 11 to float data type
second_df[column_names[3:11]] = second_df[column_names[3:11]].astype(float)
```

```python
#Changing Celceius to Farenheit
farenheit = lambda c: (c * (9/5)) + 32
second_df['Temp'] = second_df['Temp'].apply(farenheit)
```

```python
second_df
```

```python
#Changing year from string into timeseries
#Must convert year(int) to str
second_df['year'] = second_df['year'].astype(str)
year_to_dt = lambda d: datetime.datetime.strptime(d,'%Y')
second_df['year'] = second_df['year'].apply(year_to_dt)
second_df.set_index('year',inplace=True,drop=True)
```
